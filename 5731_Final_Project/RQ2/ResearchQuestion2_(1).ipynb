2. Analyze Coverage and Comprehensiveness

To analyze how well the summaries cover key concepts and terms, you can use the following script. It calculates the frequency of 'top_terms' in both types of summaries:

```python
# Print out a few examples of key_terms from the dataset
print(data['top_terms'].head())
```

```python
import ast

def calculate_term_coverage(summary, top_terms):
    # Splitting the comma-separated string into a list of terms
    terms = top_terms.split(", ")
    coverage = sum(term in summary for term in terms)
    return coverage


# Calculate coverage for each summary type
data['extractive_coverage'] = data.apply(lambda row: calculate_term_coverage(row['extractive_summary'], row['top_terms']), axis=1)
data['abstractive_coverage'] = data.apply(lambda row: calculate_term_coverage(row['abstractive_summary'], row['top_terms']), axis=1)

# Display the coverage results
print(data[['extractive_coverage', 'abstractive_coverage']].describe())
```

2. Keyword Coverage Metric

You've already created the script for calculating keyword coverage in the previous section. You can use the same script to incorporate this metric into your evaluation.

```python
def calculate_term_coverage(summary, top_terms):
    # Splitting the comma-separated string into a list of terms
    terms = top_terms.split(", ")
    coverage = sum(term in summary for term in terms)
    return coverage

# Calculate coverage for each summary type
data['extractive_coverage'] = data.apply(lambda row: calculate_term_coverage(row['extractive_summary'], row['top_terms']), axis=1)
data['abstractive_coverage'] = data.apply(lambda row: calculate_term_coverage(row['abstractive_summary'], row['top_terms']), axis=1)

# Display the coverage results
print(data[['extractive_coverage', 'abstractive_coverage']].describe())
```

4. Bar Chart of Average Readability Scores

A bar chart to compare the average readability scores between the original texts and the summaries.

Calculate Readability Scores for Original Texts

First, calculate the readability scores for the original texts:

```python
import textstat
import matplotlib.pyplot as plt

# Calculate the readability scores for 'Abstract', 'Claims', and 'Description'
data['readability_abstract'] = data['Abstract'].apply(textstat.flesch_reading_ease)
data['readability_claims'] = data['Claims'].apply(textstat.flesch_reading_ease)
data['readability_description'] = data['Description'].apply(textstat.flesch_reading_ease)

# Calculate the average readability score for the original text (including 'Abstract', 'Claims', and 'Description')
data['readability_original'] = (data['readability_abstract'] + data['readability_claims'] + data['readability_description']) / 3

# Calculate the average readability score for the 'Summary' column
average_readability = data[['readability_original', 'readability_summary']].mean()

# Plotting the average readability scores
average_readability.plot(kind='bar')
plt.title('Average Readability Scores')
plt.xlabel('Type')
plt.ylabel('Readability Score')
plt.xticks(ticks=[0, 1], labels=['Original', 'Summary'], rotation=0)
plt.show()

```
